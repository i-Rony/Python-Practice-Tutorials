```python
# x2 + y2 = 36     #circle
# x2/2 + y2/1 = 1   #ellipse

from numpy import *
import math
import random
from random import randrange
import matplotlib.pyplot as plt 
```

```python
def fn1(x):
  return round(math.sqrt(36-math.pow(x,2)), 4)
```

```python
x = []
y = []

for i in range(200):
  eachval = []
  eachval.append(round(random.uniform(-6, 6), 4))
  rval = randrange(1,3)
  if rval == 1:
    eachval.append(fn1(eachval[0]))
  else:
    eachval.append(-fn1(eachval[0]))

  x.append(eachval)
  y.append(0)

print(x)
print(y)

plt.scatter([row[0] for row in x], [row[1] for row in x], c='r', label = "Circle") 
```

```python
def fn2(x):
  return round(2*math.sqrt(1-(math.pow(x,2)/16)), 4)
```

```python
x1 = []
y1 = []
ctr = 0

while ctr < 200:
  eachval = []
  eachval.append(round(random.uniform(-6, 6), 4))
  rval = randrange(1,3)
  try:
    if rval == 1:
      eachval.append(fn2(eachval[0]))
    else:
      eachval.append(-fn2(eachval[0]))
  except:
    continue

  x1.append(eachval)
  y1.append(1)
  ctr += 1

print(x1)
print(y1)

plt.scatter([row[0] for row in x1], [row[1] for row in x1], label = "Ellipse") 
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter([row[0] for row in x], [row[1] for row in x], s=10, c='r', label='Circle')
ax1.scatter([row[0] for row in x1], [row[1] for row in x1], s=10, c='b', label='Ellipse')
plt.legend(loc='upper left');
plt.show()
```

```python
import numpy as np
from sklearn import preprocessing
from sklearn import svm

X_train = x[:139] + x1[:139]
Y_train = y[:139] + y1[:139]

X_test = x[140:] + x1[140:]
Y_test = y[140:] + y1[140:]

clf = svm.SVC()
clf.fit(X_train, Y_train)

#print(clf.feature_importances_)
test_result = clf.predict(X_test)

print("Results obtained: ", test_result)
correct = 0
for i in range(len(test_result)):
  if test_result[i] == Y_test[i]:
    correct += 1

print('Accuracy using SVM (70-30) = ', (correct*100)/len(test_result))
```
