```python
# Essential Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Preparation
X = []
Y = []

for line in open('Desktop/data_1d.csv'):
    x, y = line.split(',')
    X.append ( float(x) )
    Y.append ( float(y) )
    
X = np.array(X)
Y = np.array(Y)

# Data Visualization
plt.scatter(X,Y)
plt.show()
```

```python
# Model Generation
denominator = X.dot(X) - X.mean() * X.sum()

a = (X.dot(Y) - Y.mean() * X.sum()) / denominator
b = (Y.mean() * X.dot(X) - X.mean() * X.dot(Y)) / denominator

# Model Prediction
Yhat = a*X + b

# Model Results
plt.scatter(X,Y)
plt.plot(X,Yhat)
plt.show()
```
